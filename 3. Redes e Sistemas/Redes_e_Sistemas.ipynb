<h1 style="color:cornflowerblue">Redes e Sistemas</h1>

<h2>O Que São Redes</h2>
<p>
Redes são conjuntos de dois ou mais dispositivos eletrônicos conectados entre si, que trocam informações por meio de uma linguagem pré-estabelecida chamada protocolo.<br>
Essa conexão pode ocorrer de duas formas:
<li>Conexão Física (cabos de par trançados, coaxial ou Fibra Ótica)
<li>Via Wireless (Rádio frequência, bluetooth, infravermelho...)<br>
São os protocolos e padrões estabelecidos que garantem a correta comunicação entre dispositivos, independente de sua fabricação e localização.
</p>

<h2>Infraestruturas de Redes e os Principais Equipamentos</h2>
<p>
<li>NIC - Network Interface Card<br>
Placa de rede, como é mais conhecida, é o dispositivo responsável por permitir a conexão do computador com o cabo de rede Ethernet ou por receber as ondas de rário frequência das conexões wireless. Possui dois endereços, o Físico (MAC) e Lógico (IP).<br>
<br>
<li>HUB<br>
É um hardware que permite a conexão entre dispositivos através de cabos de par trançados e conectados entre si. Não é mais utilizado, ideia do HUB era a de ser um centralizador de equipamentos, pelas entradas de RJ45, de maneira semelhante a NIC. Dessa forma os computadores conectados no HUB poderiam conversar entre si. Mas o HUB não permite gerir canais, então a comunicação é sempre de um elemento conectado para todos os demais. O HUB não faz conexões diretas (UNICAST), apenas conexões em grupo (BROADCAST).<br>
<br>
<li>Switch<br>
Responsável pela comutação (troca ou encaminhamento da informação) dos quadros entre os dispositivos. Veio para substituir o HUB, por conta de permitir uma conexão ponto a ponto. Possui uma quantidade maior de porntas (24 ou 48, usualmente), permitindo escalabilidade da rede.<br>
<br>
<li>Roteador<br>
Tem a responsabilidade de procurar as nelhores rotas na internet para entregar os pacotes do remetente ao destinatário no menor tempo possível.<br>
<br>
<li>Modem<br>
É responsável pela modulação e a demodulação do sinal de internet. Através de um provedor de internet (ISP), o canal de internet é provido para a casa, para ser conectado aos computadores. Por vezes esse modem faz papela de modem, roteador e até switch em aparelhos domésticos.<br>
<br>
</p>

<p>
São padrões estabelecidos que definem como serão as organizações dos cabos e seus periféricos possibilitando melhor organização e performance na rede. Algumas normas definem quais devem ser os cabos, a organização, a velocidade alcançavél, entre outros.
<ul>
<li>Normas
    <ul>
    <li>NBR 14.565
    <li>ANSI/TIA 568
    <li>ANSI/TIA 569
    </ul>
</ul>
Alguns tipos de Cabos:
<ul>
<li>Par Trançado:
    <ul>
    <li>São divididos em categorias que determinam a velocidade de transmissão dos dados e o alcance em metros que o cabo pode suportar sem perda de pacotes.
    <li>UTP: Cabo sem isolamento entre os fios.
    <li>STP: Cabo com isolamento e malha entre os fios.
    </ul>
<li>Cabo Coaxial:
    <ul>
    <li>É composto por fios de cobre, tendo um fio central responsável por ser o condutor do pulso elétrico, malha metálica realizando isolamento e uma blindagem plástica contra interferências externas.
    <li>É muito utilizado em conexões com modem, com telefone e a entrega do sinal do ISP em residência por TV a cabo, por exemplo.
    </ul>
<li>Fibra Ótica:
    <ul>
    <li>É a tecnologia guiada por cabo que oferece a maior velocidade de transmissão de dados chegando a altas velocidades na casa dos Gbps. A fibra é composta por pedaços de vidros que permitem a propagação dos raios de luz que são convertidos por conversores nas extremidades das fibras. A idéia geral é que um sinal de luz é disparado em uma ponta, replicado pelos pedaços de vidro até chegar na outra extremidade. Por se tratar de luz, a velocidade é muito grande. Também é importante por conta dos cabos submersos que conectam os países e continentes (backbone).
    </ul>
<li>Rack:
    <ul>
    <li>O rack é um armário para hospedar os equipamentos de hardwares como switches, roteadores, modens, fibras ópticas e organizar os cabos em path panels. Ele é essencial em data centers e infraestruturas de redes.
    </ul>
</ul>
</p>

<h2>Modelo OSI e TCP/IP</h2>
<p>
Para a comunicação entre todos os dispositivos e estruturas acontecerem, é preciso de normas e padrões de comunicação. Esse protocolo é o modelo OSI (Open System Interconnection) ou o modelo TCP/IP (referencia aos endereços de envio e de recebimento).

<li>Modelo OSI:
    <ul>
    <li>A rede é segmentada em 7 camadas, sendo elas:
        <ul>
        <li>Aplicação (7) - Camada mais próxima dos usuários. É a camada dos procotolos DNS, SSH.
        <li>Apresentação (6) - Recebe o dado da Aplicação/Sessão. É onde está a criptografia dos dados e a segurança da informação.
        <li>Sessão (5) - É a camada responsável por estabelecer uma sessão entre o usuário e o destino e as característica dessa sessão.
        <li>Transporte (4) - É responsável pela conexão do destino, através dos protocolos TCP e UDP.
        <li>Rede (3) - É onde acontece o envio dos dados entre pontos, roteadores, etc. Nessa etapa os dados já não são segmentos, mas sim pacotes de dados, por terem acréscimos de informação.
        <li>Enlace(2) - É responsável por fragmentar o pacote de dados em quadros, que serão enviados para o dispositivo de destino. Faz também a utilização dos endereços físicos (MAC address) para envio dos dados.
        <li>Física (1) - É toda a parte física, desde cabos, roteadores, switches e todos os periféricos que fazem parte da conexão. Existem protocolos que definem as regras de como a rede fsísica transmite dados e informação (Ex: protocolo de Ethernet). 
        </ul>
    <li>A comunicação pode ser tanto no sentido 7 -> 1 (Enviando Requisição) como em 1 -> 7 (Recebendo Requisição)
    <li>Nem toda comunicação passa por todas as camadas do modelo OSI.
    </ul>

<li>Modelo TCP/IP:
    <ul>
    <li>A rede é segmentada em 4 camadas, sendo elas:
        <ul>
        <li>Aplicação (4) - Equivalente às camadas Aplicação (7), Apresentação (6) e Sessão (5) do modelo OSI.
        <li>Transporte (3) - Equivalente à camada Transporte (4) do modelo OSI.
        <li>Internet (2) - Equivalente à camada Rede (3) do modelo OSI.
        <li>Acesso a Rede (1) - Equivalente às camadas Enlace (2) e Física (1) do modelo OSI.
        </ul>
    <li>A comunicação pode ser tanto no sentido 4 -> 1 (Enviando Requisição) como em 1 -> 4 (Recebendo Requisição)
    </ul>

<li>Protocolo TCP:<br>
Envia um dado para o destino. Esse dado é nomeado como "segmento", com informações do que está sendo enviado e para onde está sendo enviado. Ao enviar, o TCP verifica a confirmação do recebimento dessa informação com o destino (e em caso negativo faz o reenvio).<br>
<br>

<li>Protocolo UDP:<br>
Semelhanter ao TCP, mas não faz a confirmação do recebimento da informação.<br>
<br>
</p>

<h2>IPV4 e IPV6</h2>
<p>
Protocolos de IP são os mais utilizados nas redes de computadores e são muito importantes para a tudo acontecer. Nas camadas de Rede (OSI) e Internet (TCP/IP) existem os roteadores que fazem o direcionamento dos pacotes de dados. Isso é feito atravez do endereço IP para o qual os pacotes devem ser enviados (IP do roteador de origem e o IP do roteador de destino).<br>
O termo IP (Internet Protocol) é o protocolo responsável pelo endereçamento dos pacotes de rede na camada de Rede (OSI) e Internet (TCP/IP), ou seja, é responsável por gerar um endereço ao seu computador, ou qualquer servidor, no momento que este conecta-se à internet. Atualmente existem dois formatos IPV4 e IPV6.<br>
<br>

<li>IPV4:
    <ul>
    <li>Foi criado na década de 80, junto da origem na internet (ARPANET)
    <li>Formato: 192.168.0.1 .
    <li>Formato de 32 bits, dividido em 8 octetos onde cada octeto pode variar de 0 até 255.
    <li>
    </ul>
<br>

<li>IPV6:
    <ul>
    <li>Desenvolvido a partir das limitações do IPV4 e a quantidade de IPs dispiníveis.
    <li>Formato: 1050:0000:0000:0000:0005:0600:300c:326b (1050:0:0:0:5:600:300c:326b).
    <li>Formato de 128 bits dividido em 16 pares.
    <li>Tem a vantagem de poder ser feita uma conhexão direta pelos IPs públicos de IPV6, sem necessidade de NAT.
    </ul>
</ul>
<br>

<li>Bit vs Byte:<br>
A comunicação pela internet ocorre através de pulsos elétricos (rede ou ondas de rádio frequência), que são recebidas e converter. Sendo assim:
    <ul>
    <li>Bit: todo o pulso elétrico - 1 significa positivo ou ligado, 0 significa desligado ou negativo;
    <li>Byte: conjunto de bits. Necessários 8 bits para formar 1 byte. Um conjunto de bits é necessário para formar um caractere que pode ser visível.
    <li>
    </ul>

<br>
<li>NAT (Network Address Translation)<br>
Por conta dos endereços de IPV4 já terem estourado a capacidade, a tecnologia NAT permite a continuidade da utilização.<br>
A partir dela, os roteadores, conectados em uma determinada rede, utilizam do IP público adicionado de uma hash para identificação do campo de origem.<br>
</p>

<h2>Cálculo de Sub Rede</h2>


<p>
Por conta das limitações da IPV4, além do NAT, foi criado o conceito de subnet para lidar com a extrapolação de endereços de rede. A subnet, ou sub redes, seria a segmentação da rede local, de modo que, por exemplo, a rede de um departamento de uma empresa não seja acessada por um outro departamento. Cada subrede vai ter um range dos IPs disponíveis para elas e a comunicação se dá apenas com rotas criadas. Se dois computadores não estão em uma rota pré estabelecida, eles não possuem comunicação.<br>
<br>

<li>Quality of Service:<br>
Define, entre outras palavras, o quanto uma rede ou a banda de uma rede, deve estar disponível para os computadores, baseados na necessidade dos serviços.<br>

Nesse caso, 192, são os números 128 e 64. Coloca-se então o número 1 na casa referente ao expoente de cada número, de acordo com a base 2:<br>
2^7 2^6 2^5 2^4 2^3 2^2 2^1 2^0<br>

<li>Máscara de Rede<br>

<br>

<li>Broadcast: endereço para mandar um pacote de dados para todos os endpoints da rede. Por convensão é o último endereço da rede.<br>

<h2>Domínio, DNS e Latência</h2>
<p>
DNS (Domain Name System), é responsável por fazer o mapeamento de um endereço familiar (google.com.br, por exemplo) e o direcionamento para o IP Público daquele endereço.<br>
O serviço DNS funciona através de servidores em uma hierarquia qe inicia pelos servidores raíz (Root) que direciona para os servidores subordinados até chegar no destino final. Ex:<br>
<li>1- Servidor Raíz - Identifica o domínio e passa para o servidor correspondente
<li>2- TLD (Top Level Domain).org .com .net - O servidor do domínio passa para o endereço de IP
<li>3- Servidor de destino: wikipedia.org google.com speedtest.net - O servidor pode ou não passar para mais uma servidor
<li>4- Servidor de destino: en.wikipedia.org mail.google.com beta.speedtest.net<br>
<br>

Para caracterizar domínio e subdomínio, segue o exemplo:
<li>https://www.umbler.com/br
    <ul>
    <li>https:// -> Protocolo
    <li>www. -> Subdomínio
    <li>umbler -> Domínio
    <li>.com -> TLD
    <li>/br -> Subdiretório
    </ul><br>
<br>
Latência: é o tempo de duração da requisição, da origme até o destino. Uma latência baixa, significa uma velocidade rápida de conexção, enquanto uma alta tem problemas de lentidão.<br>
Para diminuir a latência, alguns sites utilizam de cache, ou seja: Ao realizar uma requisição, parte dos dados do site, para que não seja necessário que a requisição seja feita até outro país novamente. Funciona da seguinte forma:
<li>Faço a requisição de uma aplicação nos EUA
<li>A minha aplicação mandou informação para a origem e colocou essa informação no cache.
<li>O cache armazenado é mais próximo (CDN)<br>
<br>
O CDN (Content Delivery Network) se trata de uma rede de servidores que vão no servidor raiz e replicam e armzenam cache. Dessa forma, a latência é diminuida pela proximidade com os CDN que com o endereço original.<br>
