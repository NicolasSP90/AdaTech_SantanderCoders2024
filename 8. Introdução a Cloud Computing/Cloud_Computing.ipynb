<h1 style="color:cornflowerblue">Introdução a Cloud Computing</h1>

<h2>Apresentação e Histórico</h2>
<p>
Os primórdios de cloud computing iniciaram em 1960 a partir de conceitos de compartilhamento de recursos computacionais que em certos momentos ficavam ociosos e até desperdiçados.<br>
Em cerca de 1997, em Atlanta, inicia-se o conceito de computação utilitária, ou seja, utilizar um recurso computacional conforme sua demanda, sendo alocado o recurso apenas em um determinado período, e o pagamento apenas dos recursos utilizados.<br>
Dessa forma, existe um recurso compartilhado e que é acessado a partir da internet, sem limite de usuários, sem limites físicos e com pagamento sob demanda. Essa é a ideia de Cloud Computing.
</p>

<h2>Cloud Pública x Privada e Principais Provedores</h2>
<p>
Antes de entrar nos conceitos de Cloud Pública e Privada, deve-se entender o conceito de DataCenter.<br>
Um Datacenter é um centro de processamento de dados. Normalmente se encontra afastado de grandes cidades, possui redundâncias de energia elétrica, internet, sistemas de resfriamento. A idéia do datacenter é comportar grandes sistemas de computadores.<br> 
Por todos os aspectos, manter e administrar um datacenter é complexo e custoso.<br>
<br>

<li>Hypervisor<br>
A partir de uma estrutura (storages, servidores, bancos de dados) é instalado o sistema Hypervisor, para permitir a virtualização de recursos. Dessa forma, são criadas máquinas virtuais que utilizam os recursos da estrutura.Ex: se minha estrutura tem uma capacidade de 80GB de RAM, posso fazer máquinas virtuais com 10, 10, 15, 20 e 25GB de RAM para atuar sob demanda.<br>
<br>

<li>Cloud Privada<br>
Possui uma estrutura de um datacenter, próprio ou alugado de terceiro, e ali é montada a estrutura cloud.<br>
É feita montagem e virtualização dos recursos para que sejam oferecidos sob demanda.<br>
Principais fornecedores: redhat e vmware.<br>
<br>

<li>Cloud Pública<br>
Possui grandes provedores: AWS, Azure, Google Cloud Plataform, Oracle, IBM, Alibaba (Oriente), Huawei (Oriente), Tencent (Oriente). A idéia nesse caso é não ser necessário manter o datacenter, mas focar nas aplicações e serviços que vão ser realizados e contratar a alocação desses recursos em datacenters de outras empresas.<br>
<br>

<li>Cloud Híbrida<br>
Quando se tem um datacenter, gerenciamento de recursos e ele se conecta da cloud privada com a cloud pública e é oferecida a aplicação utilizando as duas pontas.<br>
<br>

</p>

<h2>Principais serviços - Parte 1</h2>
<p>
Para esta aula, serão utilizados os serviços da AWS, por ser uma das principais empresas no mercado. Além disso, será utilizada pela facilidade e referência nesse contexto de cloud.<br>
Antes mesmos dos serviços é importante destacar alguns conceitos que são importantes e precisam ser pensados:<br>
<li>Localização dos data centers: É imporante verificar onde são as regiões que mais demandam o serviço, para garantir a menor latência nos serviços prestados. Além disso, por exemplo, exite a LGPD que dita que muitos dados não podem ser armazenadas fora do Brasil.
<li>Zonas de Disponibilidade: em quantos pontos a estrutura está dividina na região. É importante pensar que esses pontos podem servir de redundância das informações, fazendo com que se tenha mais segurança na disponibilidade dos serviços, caso ocorra algum problema em um dos pontos.

<li>Aba de Billing ou Preferencia de Faturamente<br>
Configurar para receber avisos sobre faturamento.<br>
<br>

<li>ClowdWatch<br>
Configurar para receber alarme para custos na plataforma.<br>
<br>


<li>IAM<br>
Gerenciador de usuários e permissões em Cloud Computing.<br>
No painel de IAM as Policies (Políticas) são informações em json com as permissões de acesso a recursos. A Role (Função) são conjuntos de Policies que permitem ou não algo.<br>
<br>

<li>EC2<br>
Serviço de provisionamento de máquinas virtuais (instancias). Como os servidores funcionam.<br>
Enquanto se cria instâncias, a AWS oferece um conceito de AMI (Amazon Machine Image), onde a instância é criado com algumas opções de sistema operacional.
<br>

</p>

<h2>Principais serviços - Parte 2</h2>
<p>
Serão visualizados serviços para fazer o armazenamento de recursos e informações.

<li>RDS<br>
Amazon Relationa Database Service: https://aws.amazon.com/pt/rds/<br>
Documentação: https://docs.aws.amazon.com/AmazonRDS/latest/UserGuide/Welcome.html <br>
O RDS é um PaaS, ou seja, a Amazon vai provisionar o recurso, a configuração de rede, armazenamento e OS. Só é necessário o inserir o esquema de banco de dados, os scripts e o gerenciamento de banco de dados.<br>
São disponibilizados bancos de dados para utilização, Open-Source (PostgreSQL...), Proprietários (Windows Server, Oracle...)<br>
<br>

<li>DynamoDB<br>
Pode ser acessado pelo painel da AWS. A documentação está na mesma página.<br>
É um serviço de banco não relaciona (NoSQL). São tabelas sem dependência entre elas e sem um esquema definido de como os bancos devem funcionar.<br>
Por mais que não siga uma estrutura, é necessário colocar uma chave de identificação única para poder fazer o gerenciamento de recursos.<br>
<br>

Tanto o RDS como o DynamoDB são escaláveis, resilientes e performáticos, podendo ser utilizados para todas as aplicações.

<li>S3<br>
Permite fazer o armazenamento de qualquer tipo de arquivo.<br>
Ao criar um bucket (estrutura para armazenamento), o nome dele é único em toda a AWS. Necessário se atentar a isso.<br>
</p>

<h2>Arquitetura para Cloud</h2>
<p>
De modo semelhante a um prédio, para um serviço de Cloud ser considrado uma boa solução e que utilize da melhor maneira possível os recursos de Cloud Computing, a arquitetura do serviço precisa ser muito bem definida, de modo que utilize os melhores recursos, as melhores ferramentas e pagar o menor valor possível.<br>
A documentação pode ser acessada em: https://aws.amazon.com/pt/architecture/well-architected/<br>
Também é importante a leitura: https://wa.aws.amazon.com <br>
<br>

<li>Well Architected Framework<br>
É um framework para arquitetura cloud com boas práticas de modo que assegure:
    <ul type="circle">
    <li>Excelência Operacional
    <li>Segurança
    <li>Confiabilidade
    <li>Eficiência
    <li>Otimização de Custos
    <li>Sustentabilidade.
    </ul><br>

Pontos importantes que devem ser pensados:
<li>Resiliência<br>
É importante que a aplicação consiga se recuperar e voltar ao seu estado natural de forma automática e sem intervensão manual. A ASE possui o serviço FIS (Fault Injection Simulator), onde é possível testar falhas no serviço, consumo de memória, testes de estress, entre outros.<br>
<br>

<li>Escalabilidade<br>
A aplicação deve conseguir escalar de forma autônoma, conforme a demana. Isso pode ser monitorado através de avisos, métricas e alarmes. A AWS tem configurações de auto-scaling que podem ser configuradas de acordo com alguns templates configuráveis.<br>
<br>

<li>Performance<br>
A aplicação deve ter uma latência baixa. É importante que a aplicação deve ser performática, conseguindo executar todos os processos e tarefas no menor tempo possível e da melhor forma.<br>
<br>

</p>

<h2>Introdução à Infraestrutura como Código</h2>
<p>
A infraestrutura pode ser setada com a interface da AWS, mas ela também pode ser gerenciada com códigos (IaaC). Dessa forma, é possível versionar, controlar melhor e padrionizar a estrutura, além de ser possível auditar.

<li>Terraform<br>
Um dos principais players. É agnóstico na sua concepção, podendo ser utilizado na AWS, Azure, etc, e seus recursos podem ser alocados em vários provedores ao mesmo tempo. Através dele, os processos são criados e executados pelos provedores.<br>
<br>

<li>Cloudformation<br>
Não é agnóstico. É exclusivo da AWS. Possui uma sinergia maior com os produtos da AWS, mas é vinculado apenas a AWS e pode ser acessado pelo painel da AWS.<br>
